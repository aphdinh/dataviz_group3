```python
customers_per_country_per_center = customer_country.groupby(["CustomerCountry","PlantKey"])["CustomerKey"].count().reset_index()
customers_per_country_per_center
```

```python
customerkey_country
```

```python
sales_new = sales_new.merge(customerkey_country, on="CustomerKey")
sales_new
```

```python
sales_new.rename(columns={'PlantKey': 'distribution_center', 'CustomerCountry':'country'}, inplace=True)
sales_new
```

```python
country_orders = sales_new.groupby("country")["SalesOrder"].count().reset_index()
country_orders.columns = ["country","total_orders_country"]
country_orders
```

```python
center_orders = sales_new.groupby("distribution_center")["SalesOrder"].count().reset_index()
center_orders.columns = ["distribution_center","total_orders_center"]
center_orders
```

```python
country_sales = sales_new.groupby("country")["OrderQuantity"].sum().reset_index()
country_sales.columns = ["country","total_sales_country"]
country_sales
```

```python
center_sales = sales_new.groupby("distribution_center")["OrderQuantity"].sum().reset_index()
center_sales.columns = ["distribution_center","total_sales_center"]
center_sales
```

```python
order_delivery = sales_new[["RequestedDeliveryDate","DeliveryDate","country","distribution_center"]]
order_delivery
```

```python
order_delivery.dtypes
```

```python
order_delivery['RequestedDeliveryDate'] = pd.to_datetime(order_delivery['RequestedDeliveryDate'])
order_delivery['DeliveryDate'] = pd.to_datetime(order_delivery['DeliveryDate'])
```

```python


new = order_delivery.loc[475245,"RequestedDeliveryDate"] - order_delivery.loc[475245,"DeliveryDate"]
new.days
if (new.days == 0) | (new.days > 0):
    print("On Time")
else:
    print("Delayed")
```

```python
order_delivery["delivery_dates"] = list(zip(order_delivery["RequestedDeliveryDate"], order_delivery["DeliveryDate"]))
```

```python
def on_time(x):
    dif = x[0]-x[1]
    if dif.days >= 0:
        return True
    else: 
        return False
    
def delayed(k):
    dif_2 = k[0] - k[1]
    if dif_2.days < 0:
        return True
    else:
        return False
```

```python
order_delivery["order_on_time"] = order_delivery["delivery_dates"].apply(on_time)
order_delivery["order_delayed"] = order_delivery["delivery_dates"].apply(delayed)
```

```python
order_delivery.loc[3,"delivery_dates"][1] - order_delivery.loc[3,"delivery_dates"][0]
```

```python
order_delivery
```

```python
delayed_orders_country = order_delivery.groupby("country")["order_delayed"].sum().reset_index()
delayed_orders_country.columns = ["country","order_delays_country"]
delayed_orders_country
```

```python
on_time_country = order_delivery.groupby("country")["order_on_time"].sum().reset_index()
on_time_country.columns = ["country","order_on_time_country"]
on_time_country
```

```python
final_sales = sales_new.merge(delayed_orders_country, on="country") \
                            .merge(on_time_country, on="country") \
                            .merge(country_orders, on="country") \
                            .merge(center_sales, on="distribution_center") \
                            .merge(center_orders, on="distribution_center") \
                            .merge(country_sales, on="country")
final_sales 
```

```python
final_sales.drop(["SalesOrder","OrderQuantity","RequestedDeliveryDate","DeliveryDate","CustomerKey"],axis=1,inplace=True)
```

```python
final_sales = final_sales.merge(final_customers, on="country")
final_sales
```

```python
final_sales.drop("distribution_center_y", axis=1,inplace=True)
```

```python
final_sales
```

```python
final_sales["percentage_delays_country"] = round(final_sales["order_delays_country"] / final_sales["total_orders_country"],4)*100
final_sales["percentage_on_time_country"] = round(final_sales["order_on_time_country"] / final_sales["total_orders_country"],4)*100
final_sales["percentage_of_center_sales_in_country"] = round(final_sales["total_sales_country"] / final_sales["total_sales_center"],4)*100
final_sales["percentage_of_center_sales_in_country"] = round(final_sales["total_sales_country"] / final_sales["total_sales_center"],4)*100
final_sales["percentage_of_center_customers_in_country"] = round(final_sales["customers_per_country"] / final_sales["customer_per_center"],4)*100
```
